##Problem 1: Smart Light Controller

```python
#Problem 1: Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # light intially off

    def turn_on(self):
        self.status = "ON"    # making light on

    def turn_off(self):
        self.status = "OFF"   # making light off

    def show_status(self):
        print(self.name + " is " + self.status)   # showing currect status


light_name = input("Enter light name: ")
action = input("Enter action (ON/OFF): ")

light1 = SmartLight(light_name)   # creating object

if action.upper() == "ON":
    light1.turn_on()
elif action.upper() == "OFF":
    light1.turn_off()
else:
    print("Invaild action")   # if wrong input

light1.show_status()
```

##Problem 2: Employee ID Card System

```python
#Problem 2: Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept   # storing departmant

    def display_id(self):
        print("----- Employee ID Card -----")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)
        print("----------------------------")   # simple display format


emp_name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
emp_dept = input("Enter Department: ")

emp1 = Employee(emp_name, emp_id, emp_dept)   # creating object

emp1.display_id()   # showing details
```

##Problem 3: Mobile Contact Record

```python
#Problem 3: Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone   # storing phone nummber

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)   # showing contact detials


contact_name = input("Enter Contact Name: ")
phone_number = input("Enter Phone Number: ")

c1 = Contact(contact_name, phone_number)   # creating object

c1.display_contact()   # calling display method
```

##Problem 4: Product Price Tag Generator

```python
#Problem 4: Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price   # storing product prize

    def display_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))   # printing in rupees format


product_name = input("Enter Product Name: ")
product_price = input("Enter Price: ")

p1 = Product(product_name, product_price)   # creating object

p1.display_tag()   # showing price tag
```

##Problem 5: Movie Rating Display System

```python
#Problem 5: Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating   # storing movie reting

    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")   # showing rating out of 5


movie_name = input("Enter Movie Name: ")
movie_rating = input("Enter Rating: ")

m1 = Movie(movie_name, movie_rating)   # creating object

m1.display_movie()   # displaying movie details
```
